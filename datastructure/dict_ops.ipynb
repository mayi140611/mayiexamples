# 字典生成

## 由两个等长的list生成字典

```python
keylist = ['a', 'b', 'c']
valuelist = list(range(3))
dict1 = dict(zip(keylist, valuelist))
dict1
```

## 字典生成式

```python
{i: 0 for i in 'sbme'}
```

# 获取字典中的元素

```python
dict1['b']
```

```python
dict1['z']
```

```python
dict1.get('z', '0')
```

```python
dict1.items()
```

```python
dict1.keys()
```

```python
dict1.values()
```

## 求字典的长度

```python
len(dict1)
```

## 遍历字典中的元素

```python
D = {'x': 1, 'y': 2, 'z': 3}
for key in D:
    print(key, '=>', D[key])
for key, value in D.items():  # 方法二
    print(key, '=>', value)
for value in D.values():  # 方法四
    print(value)
```
