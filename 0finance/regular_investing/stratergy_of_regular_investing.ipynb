# 策略0.2

## 策略0.1缺点
* 不择时   
可能会造成，当价格很高了(如已经到前期高点，波段的上沿)，还在买入  
前一天以一个价格卖出后，第二天以相近的甚至更高的价格买入

* 买卖频繁  
* 没有考虑定投组合的相关性，导致组合的抗风险能力差，回撤大
* 没有考虑标的所处的位置，都是从零开始投  
如标的已经从底部开始上涨，如果从0开始投，  
    * 那么随着标的上涨，平均成本一天天太高；
    * 而且由于是从0开始投，本金很少，绝对收益少


## 策略0.2
仓位分为初始仓位和定投
### 初始仓位
分为4档：0-25-50-75，最高75。   
档位的选择和标的所处的位置有关：
* 标的处于高位，初始仓位选择0
* 标的处于山腰位置，初始仓位选择25-50
* 标的处于底部，初始仓位选择50-75
在选择初始仓位时，可以激进一点，比如感觉应该选择25档，那么可以考虑选择50档的可能。因为  
初始仓位一旦建立，一般不动。在以下条件下可以松动。  
如：
* 标的价格达到初始买入价格的5%，卖出初始仓位的5%；
* 标的价格达到初始买入价格的10%，再卖出5%等

以此类推……

为了保证初始仓位不会越来越少，要做到仓位回补，如：
* 当标的价格达到初始买入价格的10%，再卖出5%后，这%5的份额要在标的价额回到初始买入价格的5%的时候进行回补。

以此类推……
### 定投仓位

相比于策略0.1而言，策略0.2的定投金额比较少，一般情况下为1000  

定投起始位置：一般在标的初始买入价格下方5%开始，

定投停止位置：当标的价格达到初始买入价格后，停止定投，原则上要保证定投不会高于成本线。

定投的频率：每周最多三次

周一肯定投；如果周二开盘价小于周一，则周二投，否则周三投。第三次投情况类似  


### 标的组合配置
标的组合配置是策略0.2的核心。引入如下概念：
* 组合加权相关系数R  
这个系数用来衡量当前组合的抗风险能力。  
$$权重q=\frac{当前标的市值}{标的组合总市值}$$   
假设有n个标的，那么有$C_n^2$个标的组合，每种组合都可以求出一个相关系数
$$组合加权相关系数R=\big|\frac{(q_1q_2Corr_{12}+...+q_{n-1}q_nCorr_{(n-1)n})}{(q_1q_2+...+q_{n-1}q_n)}\big|$$
由定义可知，R的取值范围为[0, 1]， 越接近与0，抗风险能力越强!  
注:由于Corr的取值范围为[-1, 1]， 当每两个组合的Corr都等于1时，R=1。

    * 用什么来计算相关系数  
    简单收益率。简单收益率是均匀的，而对数收益率为正时会被压缩，为负时会拉伸
    * 用标的多久的数据计算相关系数  
    肯定使用近期的数据，近半年or3个月or1个月
* 组合表现指数  
这个系数用来衡量组合的表现。从哪几方面衡量组合的表现呢？
绝对年化收益率  
净投入年化收益率  
最大回撤比例  

### 如何开始策略0.2
* 通过标的流动性和近n日涨跌幅框定一批标的，可以指定1-2只标的，其它的来和指定的标的搭配为组合
* 确定标的的数量和投入的比例，计算组合加权相关系数R，从而确定出最终组合
* 标的要有角色分配，哪些是主要负责盈利，哪些是配角
* 为每个标的指定定投仓位和定投策略
