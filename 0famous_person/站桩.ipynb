# 每天站桩，你知道自己在站什么吗？

关于站桩的最早记载于《内经·上古天真中论》中，余闻上古有真人者，提携天地，把握阴阳，呼吸精气，独立守神，肌肉若一，故能寿蔽天地，无有终时，此其道生。可以说，站桩是中华民族古老的智慧，无论是养生还是技击，站桩都是其不二选择。

桩功，是中国传统拳中特有的练功方法，很多拳种种都有桩功，太极中有无极桩和浑圆桩，形意中有浑圆桩和三体式，心意拳种有经典的蹲猴桩，意拳以桩为本，桩就更多了，从插指桩、扶按桩、浑圆桩、高位扶按、钩挂、撑托桩……

对于桩功，看起来就是如此简简单单的一站，这还不容易啊，没经过传承桩功的人根本不知道桩功真正的内涵，也根本不知道自己站的对错与否，很多人这本书看一点，那个网页上凑一点，就成了自己桩功的指引。运营传统武学公众号四年多来，见了太多关于桩功的素材，也同太多所谓的桩功习练者就行交流，我可以非常负责任的告诉大家，目前百分之九十九的桩功习练者站的桩都或多或少的存在一些问题，甚至很多人存在原则性的问题。网上或者书上看到的桩功的秘籍，都是缺斤短两、被阉割得残缺不全的，很多作者自己还没有入门就开始写心得写秘籍，不少人站在半山腰上描述山顶山的风景，诸如此类的作者，我见过了太多。

对于很多桩功习练者的提问，常觉得太过于幼稚和无知，对于我给他们的回答，他们也觉得太不可思议，“怎么听天书一样？怎么在网上没见过书上没见过？怎么同谁谁谁教的不一样？……”，这些愚昧的问题搞得像我在编造谎言一样不自在。芗老说的好，传武真正的好东西，任何时候都是阳春白雪，不可能普及大众。目前练拳必谈桩、养生必谈桩，人数众多，这就像一个金字塔，绝大多数人都处在金字塔的底部，在塔顶尖的凤毛麟角。

有的读者心里想，那你来分享一些桩功实际的硬核啊？是的，接下来我会一个桩一个桩的讲，把桩里面的核心和细节都和盘托出，一定会对得起关注传统武学公众平台的粉丝，但今天先讲桩功中一些相对抽象的东西，算是桩功概论，这样首先让大家对桩功有个清晰、全面的认识，以免一开始就讲具体的桩，一叶障目，不见泰山，这比讲一个实际的桩更有价值，更符合逻辑。

站桩，首先你得明白，你站在那里干什么？

## 一、桩功三要素：结构、意念和放松

无论你站什么桩，其实就是在干三件事：结构、意念和放松，也就是准确体整如注的间架，充盈丰富的意念和在保持间架准确稳定前提下的放松，这是桩功的三要素。

1. 结构

结构，也就是桩架，是你站桩摆出来的一个样子，无论站什么桩，都得摆一个造型，如无极桩、三体式、浑圆桩、撑托桩等，每个桩的结构是不一样的，站桩如建高楼打地基，地基不深不牢，盖不起高楼，桩功不深，拳学的大厦也建不起来。

站桩摆成相应的间架，你需要常问问自己，为何有此一动？站桩的间架，是在取意，取桩意，在拳学中有言：以形取意，以意象形，意自形生，形随意转。

站桩也一样，结构出了问题，会直接导致桩意取不出来，或者取出来桩意也会出现偏差。如提插桩，小臂和十指就得垂直地面往地心插，你非要往自己大腿上插，这样提插的意感你就取不出来，浑圆桩中你的十指应该是均匀分开如持球状，手指相对，且必须在同一个平面，你非要参差不齐，上下错位，前后错位，那你十指间的弹簧或磁铁般的意感就会受到影响，桩功的第一心法“指摸手拿，舍手无他，一切尽在十之间”也就无从谈起了，关于第一心法，以后作为一个专题给大家分享。

或许有的杠头说，形无形，意无意，无意之中真意，形不是最重要的，那我告诉你，准确稳定的结构承载了技击和养生的一切，没有准确稳定的间架，你谈养生和技击没有任何意义。所谓的形无形意无意，这是高级阶段的要求，对于形已经完全熟读于心，完全可以随心所欲的驾驭，你问问自己，你已经超越形这个阶段了吗？

对于桩功结构的要求有两个：准确和稳定。

准确：间架结构，不能由着你的性子随意改变的，合理科学的间架，是多少代武术大师经验的总结，在武学上大家是不能搞修正主义的，因为你根本达不否定先辈智慧的修为，别说你，当今武术界，也找不出来，既然如此，那就好好继承前人的东西，先继承下来再说。任何一个桩架，都有它相对固定的结构，这个结构无论是养生还是技击，都是最法简效宏的。如浑圆桩的抱架，两臂的高低在一定范围都是正确的，但对手型，对两中指间距离，中指到身体中线的距离，对初学者都有相对固定的规定，也就是自己拳头两拳半左右，有人抱这圆圆的，这在意动时就出问题了；浑圆桩要求肩撑肘横，你非要把肘坠在两肋边，说这是什么沉肩坠肘，其实，你根本就不懂什么是沉肩坠肘，这样你的两肩胛骨一辈子也拉不开，意动起来，也很难找到身体混元扩大，身体左右横撕的感觉。

稳定：站桩初学者，老师都要求你固定住桩架，不能随意动，如果同一个桩第一分钟一个结构，第五分钟另一个结构，第十分钟已经面目全非了，乱动是个很麻烦的问题。站桩先让保持结构纹丝不动，是为了将来更好的动，意动一致的动。大家都知道，意动一致是武学一个很高的境界，试想，你摆好了一个间架，你没有动的意念，但你就是在偷偷的乱动，既然没有此意识，你动了间架，那就是乱动，意动不一致，说到这里，你应该明白了，为什么要保持间架结构稳定的原因了。

对间架结构的要求是体整如注，身如灌铅，站到那里，如雕像一般的稳定，这是你在桩功中意动一致的基础。

2. 意念

在武学中真正的宝贝都是以虚寓实，以假修真，正如丹田，原本没有丹田，你练成了丹田，而且练成了身体无处不丹田，这就是修行，修炼内在的东西，都离不开意念的支撑。说起意念，看不见，摸不着，但意念确实存在，而且发挥这很重要的作用，正如人有思想，但思想也是无形的东西。

在桩功中，所谓意念，就是这个桩所对应的定向思维，如果站桩不明意念为何物，那么是很麻烦的事，对站桩者而言也没什么站的价值了。站桩的意念很多，同一个桩也很多种意念，站桩不同的阶段，意念也有区别，同一个桩，身体同一个具体部位，不同老师讲的意念也可能不同，归纳起来，意念可以分类为两类，共性意念和专属意念。

共性意念，也就是所有的桩功中都可以假借的意念，如设想身体浸泡在水中，设想身体每一处都与周围环境挂满了弹簧，自己如蜘蛛一般出来弹簧中央，设想自己在风中如旗一般的接受能量，整个身体在吐纳间混元扩大和缩小，设想自己整个后背往后靠在沙发或树上的意念，尾巴骨往地心插的意念，肩撑肘横的意念…….

专属意念，是某个具体的桩所桩有的意念，在其他桩中是不具备的，如在插指桩中小臂和十指垂直地面下插的意念，在浑圆桩中就没有这种意念，在浑圆桩中抱树或纸灯笼的意念，在其他桩里就没有。

在站桩初期，绝大多数人的意念是不均衡的，也就是有的地方有，有的地方没有，有的地方强烈，有的身体部位特别淡，特别是对部分桩功习练者身体某部位曾经有过伤病的人来说，曾经受伤的部位很不敏感，意念非常淡，甚至没有，这都是很正常的，但这是不好的现象，桩功要求意念充盈和均衡，这里所说的均衡，是指身体任何部位都有，而且意念的浓度还相当。当你的意念均衡了，你的体质一定也大幅度改善了，人就是活在平衡中，内外得均衡，左右得平衡，阴阳得平衡……一旦平衡被打破，身体就出问题了。

3. 放松

放松就是身体各部位和精神思维全面处于松弛状态，但放松绝不是松懈、松垮、吊儿郎当，更不是放弃。松紧是相对的，是一对矛盾，是阴阳，形意从刚入手，先刚后柔，要练成太极似的，太极要从柔和练成刚，八卦也是刚柔相济，其他拳也是这样，对99.9%的人而言，缺的是松柔，很多人，一身硬邦邦的，这种人到死也是入不了武学之门，僵死之徒而已。

放松，是一种长期修炼的一种功夫，在桩功中，要求保持结构准备稳定的前提下，无底限的放松，不放松，你的胯就是下不去，你的肩就是架着的，你的胸也涵不好，你的呼吸也不会自然，你的气机不会启动，你的上下也难以贯通……

## 二、结构、意念和放松的关系

桩功中，结构意念和放松三者相辅相成，相互影响。结构是基础，是前提，是意念和放松的载体，结构准确稳定了，意念也会更加丰满，取出桩意也更容易，以形取意，你摆出一个三体式的形，去取浑圆桩撑抱的意这肯定不行，这例子很极端，但这能说明形对意的影响。

意念，对桩架也有支撑和扶正的作用，站浑圆桩时，老师常让你两肘放在案板上的意念，或者是两肘把扶按桩扶按的意念移植过来，老师让你屁股如坐高凳的感觉，腋下夹着馒头的感觉，这些意念，会让桩架更准确稳定。

放松一定是在结构准确的前提下的放松，不让，躺在床上最放松，这没有用，不能改造你的生理。刚开始站桩，结构不可能很准确，那尽量的准确，一点点的往准确线靠，有人说，我摆成了相对准确的结构就紧张，当然，你日常的工作生活压力和不良习惯，把身体糟蹋的一塌糊涂了，让你再次回归到健康的结构上，这是痛苦蜕变，但咬牙也得坚持，这就是修行。

放松，对结构的准确稳定，也有很好的促进作用，硬邦邦的身体，很难支撑稳定均衡的间架。放松对意念的影响，可以说，僵硬直接扼杀意念。在桌子上放一颗绿豆和一颗沙子，你用松柔的手指去触摸，你能分辨出来，梗着手指去摸，就分辨不出来，放松才会让你具备充盈的意感。

在桩功中，需要把结构、意念和放松协调统一起来，站桩干的就这三件事，说起来容易，但这难倒了多少英雄好汉。

## 三、桩功的两阶段：意想和意动

大动不如小动，小动不如不动，不动才是生生不已之动。这让很多人望文生义，站桩不是不动才是生生不已之动吗？芗老给你开了个天大的玩笑，芗老的著作，跟天书一样，芗老所谓的不动，是身体最深层次细胞级的速动，外在表现为如虫子一样的蠕动，也就是东北人说的鼓涌，这个字好像敲不出来。

什么是武术？武者动也，术者手段，桩功作为武学精髓求取之法门，怎能不动呢？桩功的动，要求很高，是意动一致的动，而不是瞎动乱动。

在桩功中，很多人只知道静静的站着不动，或者前后摇摆，根本不明意动为何物，站桩分两个阶段，即意想阶段和意动阶段。

1. 意想阶段

意想阶段就是站着不动，站着不动干什么呢？也就是做准备，为意动一致做准备，准备准确稳定完美的间架结构，让身体各部位尽量能够捕捉到均衡充盈的意念，让身体各部位在结构稳定准确的前提下去放松，这是意动的前提和基础。

在站桩前期，应该吧大部分时间分配在意想，如站桩一小时，意想分配55分钟，意动仅仅可以做一个尝试，随着站桩的深入，对结构越来也稳定准确，意念也很快的捕捉到，放松状况也不错，这时可以逐渐延长意动的时间。

2. 意动阶段

意动顾名思义就是动起来，只是在桩功中有动的严格规矩，插指桩、扶按桩、浑圆桩、高位扶按桩等所有平步桩，躯干部的意动都是一致的，也就是身体左右的开合，前后的鼓荡和上下的抽拔。当然，这是初级的意动，也是最根本的，开合意动做的驾轻就熟了，再旋腾起来意动，就更有意思了。

意动时，吸气时整个身体混元扩大，头上面领着，尾闾往地下插，功夫的本源往后下鼓荡，含胸，整个脊柱向后形成一张完美的弓，呼气时，身体压缩，本源往前鼓荡，挺胸，脊柱向前形成一张完美的弓，简而言之，脊柱在意动过程中不停的做身如反弓的运动，左右半身不停的开合往复运动，只有这样，你整条脊柱才能被逐渐柔化，大龙才会形成，才会收获健康，收获发力的资本。具体到实际的桩法，以后再具体桩法中在详细介绍，不然，这篇原创写的我吃饭睡觉时间都没有了。随便提一下，意动时的呼吸，要求细慢匀长，而且一定是倒呼吸，也就是吸气时，意念从脚底起，到小腿大腿会阴再到本源，这些东西，平时在网上都看不到的，都是门里面的东西。桩功能撑筋拔骨，更能培养浩然之气，怎么没有吐纳呢？
