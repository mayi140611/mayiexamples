# 写作秘籍：好文章是怎样炼成的
https://mp.weixin.qq.com/s/O4Z5hYXxsE-XicbHkeAXQQ

每个人都希望自己能够写得一手好文章，在自媒体时代靠笔杆子获得一份稳定的收入，这也是我的梦想。从去年四月份到现在，我坚持写了一年，从一个对文字创作完全陌生的小白，到有了一点点心得，再到参加训练营，弥补了自己野路子上的一些不足，让自己的写作更加的得心应手。虽然我只是刚起步的新人，但是愿意与大家分享一下我一年来的写作经验。


## 什么样的文章才是好文章


想写出好文章来，那么就要首先知道什么样的文章才算好文章。我个人的理解，一篇好文章和一部好电影是一样的。我们可以从一部好电影是靠什么吸引观众的来理解一下一篇好文章是靠什么吸引读者的。


一部好的电影，观众对它最大的褒奖是什么呢？常用的一句话是全程无尿点。观众在观影的过程中始终兴趣盎然，注意力被电影紧紧吸引着。一篇好的文章，也是如此。必须要紧紧抓住读者的心，不让他走神，兴趣盎然的读完为快。


好电影紧紧抓住观众靠的是什么呢？首先是情节合理，铺垫演进承转起合非常顺遂。其次是情节紧凑，没有多余的镜头，悬念一个接着一个让人喘不过气来。再次是情节精彩，各种惊险元素、喜剧元素、惊悚元素层出不穷令人难忘。


同样的，一篇好的文章能够抓住读者的心，靠的也是这三个方面，首先是脉络清晰，作者要知道自己想表达什么，也知道怎么表达最顺畅。其次是内容精炼，要砍掉所有无用的废话，能用三个字说清楚的绝不用五个字。再次是金句频频，要么让人开心大笑，要么让人恍然大悟。我把这三点总结为
## 写文章的三大纪律。


### 纪律一 脉络清晰


所谓脉络，就是文章结构。结构要严密，因为作者打算写一篇文章是为了传达自己的观点，要把观点传达出去让读者看明白，作者就要明白先说什么后说什么，前有铺，后有垫，做到前后逻辑清晰合理。


好的文章结构，给人的感觉就是有条不紊，不好的文章结构，给人的感觉就是语无伦次。有条不紊，读者就会跟随作者的引领把文章看完，并且能够领会作者要表达什么。


这是作者能够写好文章的基础，因为作者的目的是为了传达自己的思想，并且希望影响到读者让读者接受自己的思想，如果读者看文章的时候感觉作者东一榔头西一榔头的没个次序，又怎么可能领会作者要表达的思想呢？


作者这一点做的不好的原因，是
#### 对自己要表达的思想没有总结提炼能力。
只有总结提炼出来了想说什么和说清楚一件事需要哪几个方面，也就是知道了要说什么，知道先说哪个后说哪个最好，然后才能进入下一步，就是怎么说的环节。


### 纪律二 内容精炼


所谓内容，就是文章叙述。叙述要通达，如果作者知道自己要说什么，但是不知道怎么说，结果说来说去也说不明白，这样的话，作者让读者理解自己接受自己的目的也不可能达到。


好的文章叙述，给人的感觉就是言简意赅，不好的文章叙述，给人的感觉就是啰里啰嗦。言简意赅，简洁明了，读者才会被文章紧紧抓住注意力，才不会在阅读文章的时候走神。


这是读者能够看完的基础，如果做不到这一点，文章像是老太太的裹脚布一样又臭又长，那么读者看了一半甚至只看了一个开头就开始对文章产生冗长的感觉，肯定会失去继续阅读下去的耐心。


作者这一点做的不好的原因，
#### 是文字表达能力欠缺，
或者是为了追求长文显得有料，刻意堆加内容，这和简洁就是天才的姐妹的理念是背道而驰的。只有用最少的话准确表达出自己，说的清之后才能进入下一步，也就是说得好的环节。


### 纪律三 金句频频


所谓金句，就是文字优化。优化要精彩，因为说的清但是说的不精彩，读者没有留下深刻的印象，就显示不出作者独树一帜的风格和水平，这样的结果是一种遗憾，因为作者只要稍微加把力，就会获得四两拨千斤的效果。


做好了文字优化，给人的感觉就是风趣幽默，没做好文字优化，给人的感觉就是刻板生硬，虽然都是用最少的话把事情说清楚了，但是有的文章让人记住了很多令人拍案叫绝的精彩语句，同样的话题文章，读者就会喜欢看你的，并对文章充满期待。这就是你区别于其他作者的竞争优势。


这里说的风趣幽默是文章优化的一个方面，我要说的意思是大家在写作中，文字表达要有特色，这种特色就像电影一样，或表现为让人轻松的搞笑，或表现为令人紧张的凝重，或表现为催人奋进的磅礴。总之一句话，文字要有力量，不能平淡。


作者这一点做不好的原因，
#### 是知识储备不足。
要想让文字生动有力量，就要活学活用各种成语典故，还有各种充满新意的变通引用，再加上各种有说服力的举例，这都需要肚子里有储备才能发挥出来。所以说，多读书，读好书，是对人生最好的投资。


## 八项注意


三大纪律总结来说，就是作者要知道自己说什么，也要知道自己应该怎么说清楚，更要知道自己怎么才能说精彩，这三点做好了，一篇好文章就有了诞生的基础，在这个基础上，作者的水平就会随着写作量的增加而日渐提高。


说完了三大纪律，接下来再说说八项注意。八项注意是每个人都会犯的小毛病，这些小毛病不像三大纪律那样一看就是硬伤，很多人甚至会把这些小毛病当做优点，所以毛病虽小，但是危害并不小，会影响作者的快速成长。


### 第一注意别违规。
不能为了快意恩仇，追求骂的痛快，而放松了警惕性。先生存，后发展，不能敌人还没打败几个呢，自己先阵亡了。我的主号被连续封了三次，一次一周一次两周一次一个月，解封之后明显感觉元气大伤，居然很快被小号超越了，同样的文章，小号能发出来，大号发不出来，所以被封是一件后果很严重的事，应极力避免。


### 第二注意别畏难。
每一个人都是从一张白纸开始变成瑰丽画卷的，起步过程就和创业过程一样，充满的是艰辛，各种失败也在所难免。很多人感觉自己写的不好就不好意思再写了，这种畏难的思想要不得，一定要鼓起勇气，多写勤写，因为不积跬步无以至千里。


### 第三注意别高冷。
很多人写作不太喜欢下力气琢磨读者想看什么，而只是想表达自己，这就容易曲高和寡没有阅读量。写的再好，如果扩散不出去，文章没有影响力，那么写文章也就失去了意义，变成了一种自娱自乐。所以大家不要刻意追求立意高，要接地气再接地气，用老百姓喜闻乐见的表达方式来写作。让自己累，不让读者累，才是好作者。


### 第四注意别贪大。
很多人写作刻意追求宏大叙事，恨不得要把能搜罗来的材料观点全都塞进去。饭要一口一口喂，不是说越多越好，一碗米饭吃着香，十碗米饭就把人撑死了。所以要懂得取舍，从小处着手，一次解决读者一个疑惑，细水长流，等能力增长了，可以驾驭宏大题材了再尝试。写作的边界是你能力的边界，而不是想象的边界。


### 第五注意押韵。
上面四个注意是从整体上来讲，具体文章写作中也有四个注意。一篇朗朗上口的文章要注意韵律感，让人读起来特别舒畅，这就要求作者在行文过程中注意哪些地方可以押韵的，在用词用字上多做斟酌。比如：天猫霸道总裁与知名网红貌似有了一腿，原配夫人一怒之下微博猛怼，眼看着三方名誉全毁，网上相关消息突然被秒删让人找不到北，直让人感觉有钱真的能使磨推鬼。这样押韵着写就容易给读者留下比较深的印象。


### 第六注意排比。
排比的合理运用可以达到一种加强语势的效果。排比的修辞功能可以概括为“增文势”“广文义”。排比项诽迭而出，语气一贯，节律强劲，各排比项意义范畴相同，带有列举和强化性质，可拓展和深化文意。比如：中国史无前例的封城闭户，美国史无前例的全民染毒，股市史无前例的暴跌一路，期货史无前例的价格为负。这样写要比平铺直叙的写更容易引发读者的兴趣。


### 第七注意对仗。
对仗又称队仗、排偶。它是把同类或对立概念的词语放在相对应的位置上使之出现相互映衬的状态，使语句更具韵味，增加词语表现力。对仗的合理运用是最容易出金句的，如果文章传播足够广，甚至有可能在互联网上永远留下你的印记。比如：忠魂不泯一腔热血化春雨，大义凛然千秋壮志泣鬼神。革命家赤胆忠心，虽死犹生；野心家祸国殃民，生不如死。这样的句子是最容易被记忆的。


### 第八注意精改。
文章写完了，不要急着在公众号上发送，要在手机上预览一下，反复阅读几遍，最好是十遍以上，看看从文章整体上还有哪些衔接连贯不完美的地方，还有哪些用字用词需要改进的地方，让整篇文章读起来有行云流水的感觉，还能顺便把所有的错别字都检查出来。自己喜欢自己满意，读者才有可能喜欢，才有可能满意，要在自己这一环节严格把关。自己感到满意了，那么恭喜你，可以发布出去让你的读者分享你的快乐了。
