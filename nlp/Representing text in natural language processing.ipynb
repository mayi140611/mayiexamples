# one-hot encoding model


## 缺点
* 它忽略了单词之间的关系，也没有传达关于单词周围环境的信息。

# N-grams language model
统计的方法  
“please eat your food”.   
A 2-gram (or bigram): “please eat”, “eat your”, or ”your food”.   
A 3-gram (or trigram): “please eat your”, or “eat your food”. 
## 用途
N-gram language models estimate the probability of the last word given the previous words. 

## 缺点
* Even with very large corpus, in general, N-gram is an insufficient model of language because language has long-distance dependencies.
* the N-gram model is heavily dependent on the training corpus used to calculate the probabilities.

# Bag-of-words language model
关心的是words在文档中的TF, 忽略word间的顺序。  


## 常用的NLP场景：  
文本分类, 如情感分析, 垃圾邮件识别;    



典型的Bag-of-words language model：  
* Naive Bayes Classifier


## 缺点
不适用于语义理解方面的NLP任务：  

* question-answering, 
* summarization, 
* dialogue
